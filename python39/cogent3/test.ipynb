```python
from cogent3 import make_tree, load_aligned_seqs
from cogent3.evolve.models import get_model
```

```python
tree = make_tree("(Galago,HowlerMon,(Rhesus,(Orangutan,(Gorilla,(Human,Chimpanzee)))));")
print(tree.ascii_art())
```

```python
aln = load_aligned_seqs("test.fasta")
```

```python
aln
```

```python
sm = get_model("HKY85")
```

```python
lf = sm.make_likelihood_function(tree, digits=2, space=3)
```

```python
lf.set_alignment(aln)
```

```python
lf.optimise()
```
